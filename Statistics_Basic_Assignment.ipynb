**Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.**

Answer:-

 Descriptive Statistics
Definition
 Descriptive statistics summarizes and describes the features of a dataset.
 It gives a snapshot of the data.


Purpose:
 To present data in a simple, understandable form without making predictions or generalizations.


Techniques Used:


Mean, Median, Mode (measures of central tendency)


Variance, Standard Deviation (measures of spread)


Tables, Graphs, Charts


Example:
A teacher calculates the average marks of students in her class.


If the average score is 75/100, that’s descriptive—it only describes her class.



2. Inferential Statistics
Definition:
 Inferential statistics makes predictions or generalizations about a population based on a sample of data.


Purpose:
 To draw conclusions or test hypotheses beyond the given data.


Techniques Used:


Hypothesis Testing (t-test, chi-square, ANOVA)


Confidence Intervals


Regression Analysis


Example:
From the marks of 50 sampled students, the teacher estimates the average score of all students in the school.


This involves inference because she is predicting about the entire population based on the sample.



**Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.**

Answer:-

SAMPLING:
Definition:
 Sampling is the process of selecting a subset of individuals (sample) from a larger group (population) to make statistical inferences about the population.


Purpose:
 Since studying an entire population is often impractical or expensive, we use samples to estimate population characteristics.


Example:
 Instead of surveying all 10,000 students in a university, a researcher may survey 500 students (sample) to estimate opinions about online classes.

Types of Sampling: Random vs Stratified
1. Random Sampling
Definition: Every individual in the population has an equal chance of being selected.


Process: Selection is purely by chance (like drawing names from a hat).


Advantage: Simple and unbiased if sample size is large enough.


Disadvantage: May not represent all groups well, especially if the population is diverse.


Example:
 Selecting 100 students randomly from a university without considering their department or year.

2. Stratified Sampling
Definition: Population is divided into subgroups (strata) based on certain characteristics (e.g., gender, age, department). Then, a random sample is taken from each subgroup.


Purpose: Ensures that all groups are proportionally represented.


Advantage: More accurate and representative, especially for diverse populations.


Disadvantage: Slightly more complex to implement.


Example:
 In a university with 60% undergraduates and 40% postgraduates, we divide students into 2 strata and then randomly select students proportionally (e.g., 60 undergraduates and 40 postgraduates for a sample of 100).





**Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.**

Answer:-


Measures of Central Tendency: Mean, Median, and Mode
1. Mean
Definition: The mean (or average) is the sum of all values divided by the total number of values.


Formula:
 Mean=Sum of all values/Number of values​

Example:
 For data: [10, 20, 30, 40, 50]

 Mean= (10+20+30+40+50)/5 = 30

2. Median
Definition: The median is the middle value when the data is arranged in ascending order.


If the number of observations is odd, median = middle value.


If even, median = average of the two middle values.


Example:
 Data: [10, 20, 30, 40, 50] → Median = 30
 Data: [10, 20, 30, 40] → Median = (20+30)/2=25



3. Mode
Definition: The mode is the most frequent value in the dataset.


A dataset can have no mode, one mode (unimodal), or multiple modes (bimodal, multimodal).


Example:
 Data: [10, 20, 20, 30, 40] → Mode = 20



Why Are They Important?
They are measures of central tendency, meaning they show the center point of a dataset.


Mean: Best for normally distributed data.


Median: Useful when data has outliers (extreme values).


Mode: Useful for categorical data (like most common brand preference).





**Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?**

Answer:-


1. Skewness
Definition: Skewness measures the asymmetry of a dataset’s distribution.


Types of Skewness:


Positive Skew (Right Skew):


Tail is longer on the right side.


Mean > Median > Mode.


Example: Income distribution (a few people earn very high salaries).


Negative Skew (Left Skew):


Tail is longer on the left side.


Mean < Median < Mode.


Example: Age at retirement (most people retire at ~60, but some retire much earlier).


Zero Skew: Symmetrical distribution (like a perfect normal distribution).



2. Kurtosis
Definition: Kurtosis measures the tailedness (heaviness of tails) of a distribution compared to a normal distribution.


Types of Kurtosis:


Mesokurtic (≈3): Normal distribution (bell-shaped).


Leptokurtic (>3): Heavy tails, more outliers (sharp peak).


Platykurtic (<3): Light tails, fewer outliers (flatter peak).



3. What Does a Positive Skew Imply?
The right tail is longer than the left.


A small number of very large values pull the mean to the right.


Most data points are concentrated on the left side of the distribution.


Example: Incomes, house prices — majority earn low/moderate amounts, but a few earn extremely high amounts.





**Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.**

**numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]**

```python
import statistics as stats   # Python's statistics module


# Given data
numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]


# Calculations
mean_value = stats.mean(numbers)
median_value = stats.median(numbers)
mode_value = stats.mode(numbers)


#For Output
print("Mean:", mean_value)
print("Median:", median_value)
print("Mode:", mode_value)
```

**Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:**

 **list_x = [10, 20, 30, 40, 50]**
 ** list_y = [15, 25, 35, 45, 60]**

```python
import pandas as pd


# Given data
list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]


# Create a pandas DataFrame
df = pd.DataFrame({'x': list_x, 'y': list_y})


# Calculate covariance
covariance = df['x'].cov(df['y'])


# Calculate correlation
correlation = df['x'].corr(df['y'])


# Output
print("Covariance between x and y:", covariance)
print("Correlation coefficient between x and y:", correlation)


```

**Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:**

**data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]**


**Answer**

     Explanation of the Result:


The boxplot visually summarizes the distribution of the data.
The box represents the interquartile range (IQR), spanning from the first quartile (Q1) to the third quartile (Q3). The line inside the box indicates the median.
The whiskers extend from the box to the minimum and maximum values within 1.5 times the IQR from the box edges.
Points outside the whiskers are considered potential outliers according to the 1.5*IQR rule.
The output from the code provides the specific values for Q1, Q3, and the IQR. It also shows the calculated lower and upper bounds for identifying outliers. Any data points falling below the lower bound or above the upper bound are listed as the identified outliers. In this case, the number 35 is identified as an outlier because it is greater than the upper bound.



```python
import matplotlib.pyplot as plt
import numpy as np


# Given data
data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


# Create a boxplot
plt.boxplot(data)


# Add title and labels
plt.title("Boxplot of Data")
plt.ylabel("Value")


plt.show()


# Identify outliers using the IQR method
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1


lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR


outliers = [x for x in data if x < lower_bound or x > upper_bound]


print("\nQ1 (25th percentile):", Q1)
print("Q3 (75th percentile):", Q3)
print("IQR (Interquartile Range):", IQR)
print("Lower Bound for outliers:", lower_bound)
print("Upper Bound for outliers:", upper_bound)
print("Identified Outliers:", outliers)
```

**Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.**

● **Explain how you would use covariance and correlation to explore this relationship.**


●** Write Python code to compute the correlation between the two lists: advertising_spend = [200, 250, 300, 400, 500] daily_sales = [2200, 2450, 2750, 3200, 4000]**


**Answer**


Explanation:
Covariance: Covariance measures the direction of the linear relationship between two variables.
A positive covariance indicates that as one variable increases, the other tends to increase as well. In this case, a positive covariance between advertising spend and daily sales would suggest that higher advertising spend is associated with higher daily sales.
A negative covariance indicates that as one variable increases, the other tends to decrease.
A covariance close to zero suggests a weak or no linear relationship.

Correlation: Correlation measures both the direction and strength of the linear relationship between two variables. The most common type is the Pearson correlation coefficient, which ranges from -1 to +1.
A correlation coefficient close to +1 indicates a strong positive linear relationship (as advertising spend increases, daily sales strongly tend to increase).
A correlation coefficient close to -1 indicates a strong negative linear relationship.
A correlation coefficient close to 0 indicates a weak or no linear relationship.
In the context of advertising spend and daily sales, we would expect a positive correlation. The closer the correlation coefficient is to +1, the stronger the evidence that increasing advertising spend leads to higher daily sales.


```python
import numpy as np


# Given data
advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]


# Convert to NumPy arrays
spend_arr = np.array(advertising_spend)
sales_arr = np.array(daily_sales)


# Compute the correlation coefficient
correlation = np.corrcoef(spend_arr, sales_arr)[0, 1]


# Output the result
print("Advertising Spend:", advertising_spend)
print("Daily Sales:", daily_sales)
print("Correlation coefficient:", correlation)
```

**Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.**

 ●** Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.**


 ●** Write Python code to create a histogram using Matplotlib for the survey data:**

 **survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]**



 **Answer**


 Explanation:


To understand the distribution of customer satisfaction scores (on a scale of 1-10), we can use the following summary statistics and visualizations:
Summary Statistics:
Mean: The average satisfaction score. This gives a central tendency of the data. A higher mean generally indicates better overall satisfaction.
Median: The middle value when the scores are ordered. The median is less sensitive to extreme values (outliers) than the mean, so it can provide a good sense of the typical satisfaction level.
Mode: The most frequent satisfaction score. This tells us which score was given by the largest number of customers.
Standard Deviation: Measures the spread or variability of the scores around the mean. A smaller standard deviation indicates that the scores are clustered closely around the mean, while a larger standard deviation suggests more spread in the satisfaction levels.
Visualizations:
Histogram: A histogram is an excellent visualization for understanding the distribution of numerical data. It shows the frequency of scores within different bins (ranges). By looking at the shape of the histogram, you can see:
If the distribution is skewed (more scores on the higher or lower end).
If there are multiple peaks (indicating different groups of customers with different satisfaction levels).
The range of scores given.
Using these statistics and visualizations together provides a comprehensive overview of customer satisfaction, helping the marketing team make informed decisions before the product launch.


```python
import matplotlib.pyplot as plt


# Given data
survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]


# Create a histogram
plt.hist(survey_scores, bins=7, rwidth=0.8, color='skyblue') # bins can be adjusted


# Add title and labels
plt.title("Distribution of Customer Satisfaction Survey Scores")
plt.xlabel("Satisfaction Score (1-10)")
plt.ylabel("Frequency")


# Set x-axis ticks to match the possible scores (1-10)
plt.xticks(range(1, 11))


# Show the plot
plt.show()
```
