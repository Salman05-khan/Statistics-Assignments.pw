**Question 1: What is a random variable in probability theory?**


**ANSWER**

A random variable in probability theory is a variable whose possible values are the outcomes of a random experiment.

-- In simple words: It is a function that assigns numerical values to the outcomes of a random process.

Discrete Random Variable

Takes countable values (finite or infinite countable).

Example: Number of heads when tossing 3 coins (0, 1, 2, 3).

Example: Height of students in a class (can be any real number within a range).

Possible values of X = {1, 2, 3, 4, 5, 6}.

**Question 2: What are the types of random variables?**



**ANSWER**


In probability theory, random variables are mainly classified into two types:


1. Discrete Random Variable

Takes countable values (finite or countably infinite).

Each value has a specific probability.

Probabilities are described using a Probability Mass Function (PMF).

 Examples:

Number of heads in 3 coin tosses → {0, 1, 2, 3}.

Number of customers in a shop in one hour.

Outcome of rolling a die → {1, 2, 3, 4, 5, 6}.




2. Continuous Random Variable

Takes uncountable values (infinite within an interval).

Probability of taking an exact value is 0; instead, probabilities are assigned over intervals.

Described using a Probability Density Function (PDF).

 Examples:

Height or weight of a person.

Time taken to complete a task.

Temperature on a given day.

**Question 3: Explain the difference between discrete and continuous distributions.**



**ANSWER**


**Discrete Distribution**

A discrete distribution is the probability distribution of a discrete random variable.

The variable can take only specific and countable values (like 0, 1, 2, 3 …).

Probabilities are assigned to each value separately, and the sum of all probabilities is equal to 1.

It is represented using a Probability Mass Function (PMF), which gives the probability of each possible outcome.

Graphically, it is shown using a bar graph with distinct bars for each outcome.

Examples:

Rolling a die (outcomes: 1, 2, 3, 4, 5, 6).

Number of heads in tossing 3 coins (outcomes: 0, 1, 2, 3).

Number of students present in a class.




**Continuous Distribution**

A continuous distribution is the probability distribution of a continuous random variable.

The variable can take any value within an interval, meaning infinitely many values are possible.

The probability of the variable taking an exact single value is zero; instead, probability is calculated over a range of values.

It is represented using a Probability Density Function (PDF), where the total area under the curve is equal to 1.

Graphically, it is shown as a smooth curve, and probabilities correspond to the area under the curve over an interval.

 Examples:

Height or weight of a person.

Time taken to complete a task.

Temperature of a city during a day.

**Question 4: What is a binomial distribution, and how is it used in probability?**


**ANSWER**


Binomial Distribution

The binomial distribution is a type of discrete probability distribution. It describes the number of successes in a fixed number of independent trials, where each trial has only two possible outcomes:

Success (with probability p)

Failure (with probability q=1-p)

If we repeat the experiment n times, the binomial distribution gives the probability of getting exactly k successes.

Formula (Probability Mass Function):


P(X=k)=(n/k​) p^k (1-p)^n-k

Where:


n = number of trials


k = number of successes


p = probability of success in one trial

1−p = probability of failure

(n/k)= n!/(k!(n−k)!) = number of ways to choose k successes out of .

Conditions for Binomial Distribution:

A fixed number of trials (n).

Each trial has two possible outcomes (success or failure).

Probability of success (p) is the same in each trial.

All trials are independent.

Examples of Usage:

Tossing a coin 10 times → probability of getting exactly 6 heads.

In quality control → probability that 3 out of 20 products are defective.

In exams → probability that a student answers exactly 7 out of 10 questions correctly if chance of success is 0.7.

**Question 5: What is the standard normal distribution, and why is it important?**


**ANSWER**


**Standard Normal Distribution**



The standard normal distribution is a special case of the normal distribution.

It is a continuous probability distribution that is symmetric and bell-shaped.

Its mean (μ) = 0 and standard deviation (σ) = 1.

The variable that follows this distribution is called the standard normal variable (Z).

The probability density function (PDF) of the standard normal distribution is:

f(z)= (1/root2π)e^((-z^2)/2),  −∞<z<∞


**Why is it important?**

Basis for many statistical methods

Most real-world continuous data (like heights, IQ scores, exam marks) approximately follow a normal distribution.

By converting them to the standard normal distribution, we can apply statistical tests and analysis easily.

Z-scores (Standardization)

Any normal random variable X∼N(μ,σ2) can be converted to the standard normal form using:

Z= (X-μ)/σ


This process is called standardization and allows comparison between different datasets.

Probability calculations

Probabilities for normal distributions are found using the Z-table (standard normal table).

Example: The probability that a Z value lies between -1 and 1 is about 68%.

Widely used in hypothesis testing

Z-tests, confidence intervals, and control charts in quality control all rely on the standard normal distribution.

Example:

If a student’s height is normally distributed with mean 170 cm and standard deviation 10 cm, the probability that the student’s height is less than 180 cm can be found by standardizing:

Z=  (180-170)/10 =1

Now, we can use the standard normal distribution to find the probability.

**Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?**


**ANSWER**


**Central Limit Theorem (CLT)**

The Central Limit Theorem states that:

When we take repeated random samples of sufficiently large size n from any population (with mean μ and finite variance σ^2), the sampling distribution of the sample mean will approximately follow a normal distribution, regardless of the original population’s shape.

Mathematically, if Xˉ is the sample mean of size

Z= (Xˉ-μ)/(σ/(root n))  ⟶ N(0,1) as n→∞


Why is it critical in statistics?

**Normal approximation**

Even if the original population is skewed or not normal, the distribution of the sample mean will be approximately normal when n is large (usually n≥30 is considered enough).


**Foundation for inferential statistics**

CLT allows us to use normal probability theory for estimation and hypothesis testing, even when we don’t know the exact population distribution.

**Works in real-life data**

Many processes in nature and society are not perfectly normal. CLT gives a way to still apply normal-based methods reliably.


**Confidence intervals & hypothesis testing**

The reason we can build confidence intervals for means and conduct Z-tests or t-tests is because of CLT.



Example:

Suppose exam scores in a school are heavily skewed, with mean 60 and standard deviation 20.

If we take one sample of 10 students, their mean score may vary widely.

But if we take many samples of 50 students each, the distribution of those sample means will form a bell-shaped curve (approximately normal), centered around 60.

**Question 7: What is the significance of confidence intervals in statistical analysis?**


**ANSWER**


**Confidence Intervals (CI)**:

A confidence interval gives a range of values within which we expect the true population parameter (like mean or proportion) to lie, with a certain level of confidence.



For example:

If we compute a 95% confidence interval for the mean exam score as [68,74], it means we are 95% confident that the true average score of the entire population lies between 68 and 74.

**Significance of Confidence Intervals in Statistical Analysis**

**Estimation with uncertainty**

Instead of giving just a single point estimate (like sample mean = 71), a CI shows the range of plausible values for the population parameter.

**Measure of reliability**

The width of the interval reflects how precise the estimate is.

A narrow CI → estimate is precise.

A wide CI → more uncertainty.

**Connection to sample size**

Larger samples produce narrower confidence intervals (more accurate estimates).

Smaller samples produce wider intervals.

**Basis for decision-making**

Used in research, surveys, medicine, and quality control to decide if results are statistically significant.



Example:

 If a 95% CI for the difference between two treatments does not include 0, we conclude the treatments are significantly different.

**Practical interpretation**

Policymakers, doctors, and businesses often prefer CIs over just p-values because they provide an actual range of possible outcomes, not just “significant or not.”

**Question 8: What is the concept of expected value in a probability distribution?**



**ANSWER**


**Expected Value (EV)**

The expected value in probability is the idea of finding the “average outcome” of a random process if it were repeated many times. It does not mean what will happen in one single trial, but rather what we should expect in the long run. That’s why it is often called the long-run average of a probability distribution.

**The Concept**

When we have a random variable (like the outcome of rolling a die or tossing a coin), each possible result has its own likelihood of occurring. The expected value combines the possible results with their probabilities and tells us where the “balance point” or “center” of the distribution lies.

For a coin toss, if you assign 1 point for heads and 0 points for tails, then over thousands of tosses the average points per toss will approach 0.5. That 0.5 is the expected value.

For a die roll, the numbers 1 to 6 all occur equally often. If you roll the die over and over, the average of all the outcomes will get closer and closer to 3.5. This is the expected value of rolling a die, even though 3.5 is not an actual result on the die.


**Why is Expected Value Important?**

Gives the central tendency – It tells us where the “average outcome” of a random variable lies.

Helps in making decisions – In gambling, business, and finance, expected value helps to decide whether a choice is favorable in the long run.

Used in risk assessment – Insurance companies use expected value to set premiums, by balancing possible payouts with the likelihood of claims.

Forms the base for further analysis – Concepts like variance and standard deviation are built upon the idea of expected value.

**Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random numbers from a normal distribution
mean = 50
std_dev = 5
num_samples = 1000
random_numbers = np.random.normal(mean, std_dev, num_samples)

# Compute the mean and standard deviation using NumPy
computed_mean = np.mean(random_numbers)
computed_std_dev = np.std(random_numbers)

print(f"Generated Mean: {computed_mean:.2f}")
print(f"Generated Standard Deviation: {computed_std_dev:.2f}")

# Draw a histogram to visualize the distribution
plt.hist(random_numbers, bins=30, density=True, alpha=0.6, color='g')
plt.title('Histogram of Random Numbers from Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.**

**daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
 235, 260, 245, 250, 225, 270, 265, 255, 250, 260]**


●**Explain how you would apply the Central Limit Theorem to estimate the average sales with a 95% confidence interval.**


● **Write the Python code to compute the mean sales and its confidence interval**




**ANSWER**



**Explanation**

So, we have 20 days of sales data, but the company wants to know the true average sales for all days in 2 years. Now, it’s not possible to check every single day, so we use the Central Limit Theorem (CLT).

The CLT basically says: if you take samples and calculate their averages, those averages will form a bell-shaped curve (normal distribution), even if the original data is not perfectly normal. This is really helpful because it means we can trust the sample mean as a good estimate of the population mean.

Here’s how we use it:

First, we calculate the average sales from the sample data — this is our best guess of the true mean.

Then, we look at how much the sales values vary — that’s the standard deviation.

Since our sample size is small (only 20 days), we use the t-distribution to calculate a 95% confidence interval.

Finally, the confidence interval gives us a range. We can say: “We are 95% confident that the true average daily sales for the company lies somewhere between the lower limit and the upper limit of this interval.”

```python
import numpy as np
from scipy.stats import t

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Calculate the sample mean
sample_mean = np.mean(daily_sales)

# Calculate the sample standard deviation
sample_std = np.std(daily_sales, ddof=1) # Use ddof=1 for sample standard deviation

# Sample size
n = len(daily_sales)

# Degrees of freedom
degrees_freedom = n - 1

# Significance level
alpha = 0.05

# Critical t-value for a 95% confidence interval (two-tailed)
t_critical = t.ppf(1 - alpha/2, degrees_freedom)

# Standard error of the mean
standard_error = sample_std / np.sqrt(n)

# Margin of error
margin_of_error = t_critical * standard_error

# Compute the confidence interval
confidence_interval_lower = sample_mean - margin_of_error
confidence_interval_upper = sample_mean + margin_of_error

print(f"Sample Mean Sales: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"95% Confidence Interval for Average Sales: [{confidence_interval_lower:.2f}, {confidence_interval_upper:.2f}]")
```
